```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
import scipy.stats as stats
import sklearn


from matplotlib import rcParams
sea.set_context("poster")
sea.set_style('whitegrid')

%pylab inline 
%matplotlib inline

df = pd.read_csv('FinishMissing.csv')
df=df.drop('Unnamed: 0',axis=1)
final=df
```

## Run OLS Regression on Categorical Features
* must run four features at a time due to memory constraints

```python
import numpy
import pandas
import statsmodels.api as sm
import seaborn
import statsmodels.formula.api as smf 
```

```python
#Run OLS on 4 categorical features at a time, memory constraints
model1=smf.ols(formula='logerror ~C(airconditioningtypeid)', data=final).fit()
model2=smf.ols(formula='logerror ~C(architecturalstyletypeid)', data=final).fit()
model3=smf.ols(formula='logerror ~C(buildingqualitytypeid)', data=final).fit()
model4=smf.ols(formula='logerror ~C(buildingclasstypeid)', data=final).fit()

print(model1.summary())
print(model2.summary())
print(model3.summary())
print(model4.summary())

```
